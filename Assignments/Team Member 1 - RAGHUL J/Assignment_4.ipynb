## 8. Scaling the data

```python
y = data["Gender"]
y.head()
```

```python
x=data.drop(columns=["Gender"],axis=1)
x.head()
```

```python
from sklearn.preprocessing import scale
X_Scaled = pd.DataFrame(scale(x), columns=x.columns)
X_Scaled.head()
```

## 9. Perform any of the clustering algorithms

```python
from sklearn.cluster import KMeans
sse = []
k_rng = range(1,10)
for k in k_rng:
    kn = KMeans(n_clusters=k)
    kn.fit(X_Scaled[['Spending Score (1-100)','Annual Income (k$)']])
    sse.append(kn.inertia_)
plt.plot(k_rng,sse)
```

```python
kn = KMeans(n_clusters=5)
y_predicted=kn.fit_predict(data[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

## 10. Add the cluster data with the primary dataset

```python
df_transformed = data
```

```python
df_transformed['class'] = y_predicted
```

```python
df_transformed
```

## 11. Split the data into dependent and independent variables.

```python
feature =pd.DataFrame(df_transformed.drop(['class'], axis = 1))
label = pd.DataFrame(df_transformed['class'])
```

```python
feature.head()
```

```python
label.head()
```

## 12. Splitting into training and test datasets.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, label, test_size = 0.2, random_state = 0)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

## 13. Build the Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10,criterion='entropy')
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict_train = model.predict(X_train)
```

## 14. Train the Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('Training accuracy: ',accuracy_score(y_train,y_predict_train))
```

## 15. Test the Model

```python
print('Testing accuracy: ',accuracy_score(y_test,y_predict))
```

## 16. Measure the performance using Evaluation Metrics.

```python
cm = confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True,cmap='viridis')
print(classification_report(y_test,y_predict))
```
